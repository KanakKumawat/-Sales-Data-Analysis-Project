# SALE BY MONTH

```python
df.groupby('Year')[[' Sales','Profit']].sum()
df.groupby('Month Name')[[' Sales','Profit']].sum()
```

```python
plt.figure(figsize=(10,3))
df.groupby('Month Name')['Profit'].sum().plot()
plt.title('Profit over the Year')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.show()
```

```python
month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
plt.figure(figsize=(10,3))
(df.groupby('Month Name')['Profit'].sum().reindex(month_order).plot())
plt.title('Profit over the Year')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.show()
```

# PROFIT BY PRODUCT

```python
df.groupby('Product')[[' Sales','Profit','Units Sold']].sum().sort_values(' Sales', ascending=False)
```

```python
plt.figure(figsize=(10,3))
df.groupby('Product')[' Sales'].sum().plot(kind='bar')
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Sales (Millions USD)')
plt.show()
```

# SALE AND PROFIT BY COUNTRY

```python
df.groupby('Country')[[' Sales','Profit']].sum()
```

```python
plt.figure(figsize=(10,3))
df.groupby('Country')[' Sales'].sum().plot(kind='bar')
plt.title('Total Sales by Country')
plt.xlabel('Country')
plt.ylabel('Sales (Millions USD)')
plt.show()
```
